```python
# Using this function
local_time = time.localtime()
year, month, day = local_time[0], local_time[1], local_time[2]
print(f"Year: {year}\nMonth: {month}\nDay: {day}")
```

```python
type(month)
```

```python
i = 0
# Store the unique time into a variable
first_time = time.time()

while i < 1000:
  if type(i) == int:
    i += 1

# Store the unique time after the loop in a variable
last_time = time.time()

# Calculate the difference
dt = last_time - first_time

print(f"{dt} seconds elapsed")
```

without if: 0.00012 seconds

with if: 0.00044 seconds

```python
import random
```

```python
# To choose a random number among the numbers in a list you can use `choice`
my_list = [1, 2, 'x', 4, 5, 6]
random.choice(my_list)
```

we're using [w3school](https://www.w3schools.com/python/module_random.asp) to check what are the other functions which we can use with random module

> We can use `random.shuffle(seq)` to shuffle a sequence and return it

```python
print(f"My list before shuffling {my_list}")
random.shuffle(my_list)
print(f"After the first shuffle {my_list}")
random.shuffle(my_list)
print(f"After the second shuffle {my_list}")
```
