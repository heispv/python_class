```python
def odd_even(my_list):
  """
  gets a list of numbers and return if they are odd or even
  """

  odd_even = []

  for num in my_list:
    if num % 2 == 0:
      odd_even.append('even')
    else:
      odd_even.append('odd')
  
  return odd_even
```

```python
def fibonacci(n):
  series = [1, 1]

  for i in range(2, n):
    new_num = series[i-1] + series[i-2]
    series.append(new_num)

  return series
```

```python
fibonacci(10)
```

```python
def fibo(n):
  lst = []
  a = 1
  b = 1

  for i in range(n):
    lst.append(a)
    a, b = b, a + b

  return lst
```

```python
fibo(10)
```
