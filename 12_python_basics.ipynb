```python
class ListeMardom:
  # class variable
  history = []
  count = 0

  def __init__(self, name):
    self.name = name

  def add(self):
    ListeMardom.history.append(self.name)
    ListeMardom.count += 1
  
  def __repr__(self):
    return self.name
```

## Inheritance

Screenshot (30).png

```python
class Computer:
  count = 0

  def __init__(self, cpu, hard, gpu):
    Computer.count += 1
    self.cpu = cpu
    self.hard = hard
    self.gpu = gpu

  def value(self):
    return self.cpu * 1000 + self.hard + self.gpu
  
```

```python
x = Computer(6, 1024, 2048)
```

```python
x.value()
```

```python
x.count
```

```python
class Laptop(Computer):
  pass
```

```python
y = Laptop(7, 1024, 4096)
```

```python
y.gpu
```

```python
y.value()
```

```python
y.count
```

```python
class Laptop(Computer):
  def __init__(self, cpu, hard, gpu, model):
    super().__init__(cpu, hard, gpu)
    self.model  = model
  
  def details(self):
    print(f"laptop e modele {self.model} daraye {self.cpu} haste va harde {self.hard} va gpu e {self.gpu} mibashad.")
```

```python
x = Laptop(6, 512, 1024, "HP")
```

```python
x.value()
```

```python
x.model
```

```python
x.gpu
```

```python
x.details()
```

```python
class LaptopHP(Laptop):
  pass
```

```python
xx = LaptopHP(4, 1024, 4096, "Lenovo")
xx.details()
```
