```python
class Adam:
  def __init__(self, name, age):
    self.name = name
    self.age = age
```

```python
y = Adam('jack', 20)
```

```python
y.name, y.age
```

```python
class Dog:
  # class variable
  bread = 'Husky'
  def __init__(self, name, age):
    # intance variable
    self.name = name
    self.age = age
```

```python
a = Dog('Taffy', 3)
b = Dog('Churuk', 5)
```

```python
a.bread, b.bread
```

```python
a.name, b.name
```

```python
a.age, b.age
```

```python
class ListeMardom:
  # class variable
  history = []
  count = 0

  def __init__(self, name):
    self.name = name

  def add(self):
    self.history.append(self.name)
    # bayad check bokonim bebinim chera injurie
    self.count += 1
```

```python
adam_1 = ListeMardom('ali')
adam_2 = ListeMardom('nazanin')
```

```python
adam_1.name, adam_2.name
```

```python
adam_1.count, adam_2.count
```

```python
adam_1.history, adam_2.history
```

```python
adam_1.add()
```
