```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-2, 2, 0.2)
y = x
y_2 = x ** 2
y_3 = x ** 3
y_4 = x ** 4
```

```python
x
```

Screenshot (26).png

```python
# In order to change the figure size
plt.figure(figsize=(10, 10))


plt.subplot(2, 2, 1)
plt.title("Linear")
plt.plot(x, y)
plt.axis("off")

plt.subplot(2, 2, 2)
plt.title("Quadratic")
plt.plot(x, y_2)
plt.axis("off")

plt.subplot(2, 2, 3)
plt.title("Cubic")
plt.plot(x, y_3)
plt.axis("off")

plt.subplot(2, 2, 4)
plt.title("Quatric")
plt.plot(x, y_4)
plt.axis("off")
```

```python
plt.figure(figsize=(10, 10))
plt.title("Different powers")
plt.plot(x, y)
plt.plot(x, y_3)
plt.plot(x, y_4)
plt.legend(("Linear", "Cubic", "Quatric"))
plt.grid()
```
